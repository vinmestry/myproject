```python
X = df.iloc[: , 2:4] # [   start_row  : stop_row , start_col: stop_col   ]
```

```python
X
```

```python
%%time 

# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
from tqdm import tqdm_notebook as tqdm

K_MAX = 9
wsse = []
# progress bar 
with tqdm(total=K_MAX-1) as pbar:
    for i in range(1, K_MAX):

        kmeans = KMeans(n_clusters = i,   random_state = 10)
        kmeans.fit(X)
        wsse.append(kmeans.inertia_)
        pbar.update(1)
```

```python
plt.plot(range(1, K_MAX), wsse)
plt.scatter(range(1, K_MAX),wsse)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WSSE')
plt.show()
```

```python
print(wsse[4])
```

```python
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 5,  random_state = 10)
kmeans.fit(X)
Y_pred = kmeans.predict(X)
```

```python
Y_pred
```

```python
kmeans.n_iter_
```

```python
df["Clusters"]=Y_pred
df.head(10)
```

```python
kmeans.cluster_centers_
```

```python
sns.lmplot(data=df, 
           x='Annual Income (k$)',
           y='Spending Score (1-100)',
           fit_reg=False, # No regression line
           hue='Clusters',
           palette="Set1")  

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
            s = 300, c = 'yellow')
plt.show()
```

```python
df["Clusters"]=df.Clusters.map({0:"Careless",1:"Sensible",2:"Target",
                              3:"NotUseful",4:"Careful"})
```
